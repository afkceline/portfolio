Como hay 7 tablas, vamos a proceder a buscar la tabla que contiene la información de la composición étnica que nos pide el ejercicio. Echamos un vistazo previo a la página web solicitada para saber qué terminos podemos aplicar para encontrar la tabla que necesitamos. En este caso, sabemos que debe contener el término 'Raza'

```python
url = 'https://es.wikipedia.org/wiki/Juegos_Ol%C3%ADmpicos'

# Simulamos un navegador para evitar el error 403
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/123.0 Safari/537.36'
}

# Realizar la solicitud GET con el User-Agent especificado
response = requests.get(url, headers=headers)

if response.status_code == 200:
    # Convertir el HTML en un archivo de html
    html_content = io.StringIO(response.text)

    # Extraer tablas sin warning
    # # Filtramos las tablas que contienen el texto "Raza" en el encabezado
    tables = pd.read_html(html_content, match='Raza')

    print(f"El número de tablas encontradas: {len(tables)} \n")
else:
    print(f"Error: {response.status_code}")
```

```python
#Como nos da error con varias palabras, haremos la búsqueda de la tabla de forma manual

display(tables[4].head())
```

```python
#Convertimos a dataframe para poder trabajar con los datos

df = tables[4]
df
```

```python
#Comprobamos el tipo de datos que tenemos en el df

df.dtypes
```

```python
#Eliminamos las columnas sin información

df_1 = df.drop(columns=['Alone.1', 'Total.1'])
df_1
```

```python
#Renombramos las columnas para que sean más claras

df_1 = df_1.rename(columns={
    'Raza y etnia[11]​': 'etnia',
    'Alone': 'porcentaje'
})
```

```python
df_1
```

```python
#Realizamos un análisis exploratorio

df_1.describe()
```

```python
#Como el anterior código nos ha arrojado resultados inesperados, volvemos a comprobar los datos

df_1.dtypes
```

```python
#Transformamos a float

df_1['porcentaje'] = (
    df_1['porcentaje']
    .str.replace('%','', regex=False)
    .str.replace(',','.', regex=False)
    .str.replace('—','', regex=False)
    .str.strip()
    .replace('', None)
    .astype(float)
)

df_1['Total'] = (
    df_1['Total']
    .str.replace('%','', regex=False)
    .str.replace(',','.', regex=False)
    .str.replace('—','', regex=False)
    .str.strip()
    .replace('', None)
    .astype(float)
)
```

```python
df_1.dtypes
```

```python
df_1[['porcentaje','Total']].describe()
```

## 🎈 3. Ejercicio beautiful soup🎈


- Beautiful Soup: scrapear la tabla de lenguajes de programación más utilizados (https://www.tiobe.com/tiobe-index/) creando un dataframe con una columna llamada "lenguaje" y otra columna llamada "porcentaje_uso" (cogiendo la info de programming language y ratings respectivamente).

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# URL
url = 'https://www.tiobe.com/tiobe-index/'

# Simulamos un navegador para evitar el error 403
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/123.0 Safari/537.36'
}

# Realizar la solicitud GET con el User-Agent especificado
response = requests.get(url, headers=headers)

if response.status_code == 200:
    print(f"status_code: {response.status_code}")
    # Convertir el HTML en un archivo de html
    soup = bs(response.text, 'html.parser')
else:
    print(f"Error: {response.status_code}")
```

```python
# soup es un tipo de dato propio de bs
# representa un HTML parseado
type(soup)
```

```python
soup.prettify()
```

```python
soup
```

```python
tables = soup.find_all('table')
```

```python
type(tables)
```

```python
len(tables)
```

```python
# Buscar la tabla específica basada en el caption
# Utilizamos soup.find_all para encontrar todas las tablas (<table>)
#  que tienen la clase 'table-top20' en el documento HTML
# filtrando por la clase se reduce el numero de coincidencias
tables = soup.find_all('table', {'class': 'table-top20'})

print(f"Hemos encontrado {len(tables)} tablas que tienen en la clase 'table-top20' en nuestro documento HTML ")

print(f"\nSi no filtráramos por la clase, encontraríamos {len(soup.find_all('table'))} tablas en nuestro documento HTML ")
```
